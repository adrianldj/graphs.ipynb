```python
from scipy.optimize import curve_fit
```

```python

gah = 65
#gah = 65, the data points that corresponds with the radius of the bulge used by dfdsf at 150 arcseconds
#gah = 54, the data points that correspond with the radius of the bulge used by S. Kendall at 46 arcseconds
```

```python

def fitFunc(test, a, b, c, d):
    return a*np.exp(-b*((semi[0:gah])**(c)-1))+d
```

```python

fitParams, fitCovariances= curve_fit(fitFunc, semi[0:gah], inte[0:gah],[19000., 1.01, 1.05, 6000.0])
print fitParams

fitParams2, fitCovariances2= curve_fit(fitFunc, semi2[0:gah], inte2[0:gah],[13000.0, 1., 1., 3000.])
print fitParams2

fitParams3, fitCovariances3= curve_fit(fitFunc, semi3[0:gah], inte3[0:gah],[13000.0, 1., 1., 5000.])
print fitParams3

fitParams4, fitCovariances4= curve_fit(fitFunc, semi4[0:gah], inte4[0:gah],[13000.0, 1., 1., 5000.])
print fitParams4
```

```python

plt.figure(figsize=(12,20))

plt.subplot(4, 1, 1)
plt.title("Sersic Profile fitting (Infrared Filter)")
plt.xscale('log')
plt.ylabel('Intensity (arbitrary units)')
#plt.yscale('log')
plt.plot(semi[0:gah],inte[0:gah],marker='o',linestyle="None",color="yellow")
plot(semi[0:gah], (fitFunc(semi[0:gah], fitParams[0], fitParams[1], fitParams[2], fitParams[3])))

plt.subplot(4, 1, 2)
plt.title("Sersic Profile fitting (Blue Filter)")
plt.xscale('log')
plt.ylabel('Intensity (arbitrary units)')
#plt.yscale('log')
plt.plot(semi2[0:gah],inte2[0:gah],marker='o',linestyle="None",color="cyan")
plot(semi2[0:gah], (fitFunc(semi2[0:gah], fitParams2[0], fitParams2[1], fitParams2[2], fitParams2[3])))

plt.subplot(4, 1, 3)
plt.title("Sersic Profile fitting (Visual Filter)")
plt.xscale('log')
plt.ylabel('Intensity (arbitrary units)')
#plt.yscale('log')
plt.plot(semi3[0:gah],inte3[0:gah],marker='o',linestyle="None",color="lime")
plot(semi3[0:gah], (fitFunc(semi3[0:gah], fitParams3[0], fitParams3[1], fitParams3[2], fitParams3[3])))

plt.subplot(4, 1, 4)
plt.title("Sersic Profile fitting (Red Filter)")
plt.xlabel('Semimajor axis (in pixels on CCD)')
plt.ylabel('Intensity (arbitrary units)')
plt.xscale('log')
#plt.yscale('log')
plt.plot(semi4[0:gah],inte4[0:gah],marker='o',linestyle="None",color="tomato")
plot(semi4[0:gah], (fitFunc(semi4[0:gah], fitParams4[0], fitParams4[1], fitParams4[2], fitParams4[3])))
```
